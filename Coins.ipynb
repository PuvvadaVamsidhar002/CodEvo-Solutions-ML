```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly
img=cv2.imread('bitss.jpg')
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img1)
plt.show()
img1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img1.shape)
plt.figure(figsize=(10,10))
plt.imshow(img1,cmap='gray')
plt.show()
```

```python
img2=cv2.GaussianBlur(img1,(5,5),0)
```

```python
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img2,cmap='gray')
plt.show()
```

```python
i,threshold1=cv2.threshold(img2,70,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(threshold1,cmap='gray')
plt.show()
```

```python
contours,hierarchy=cv2.findContours(threshold1,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)
```

```python
print('length of contours{}'.format(len(contours)))
print(contours)
```

```python
filtered_contour=[contour for contour in contours if(cv2.contourArea(contour)>10)]
count=len(filtered_contour)
```

```python
contour_image=cv2.drawContours(threshold1,filtered_contour,-1,(0,255,0),thickness=0)
contour_image_rgb = cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB)
plt.imshow(contour_image_rgb)
plt.title('Drawn Contours')
plt.axis('off')  # Hide axis
plt.show()
```

```python
contour_origin=cv2.drawContours(img.copy(),filtered_contour,-1,(0,255,0),thickness=2)
rgb_image=cv2.cvtColor(contour_origin,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(rgb_image)
plt.show()
```
